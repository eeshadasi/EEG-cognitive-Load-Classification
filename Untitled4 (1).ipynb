```python
import pandas as pd
df=pd.read_csv("all.csv")
df=df.drop(["Person","Type"],axis=1)
```

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Apply the function to each relevant column
columns_to_check = ['TP9_psd', 'TP9_de', 'AF7_psd', 'AF7_de', 'AF8_psd', 'AF8_de', 'TP10_psd', 'TP10_de']

for column in columns_to_check:
    df = remove_outliers(df, column)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras import layers, callbacks

X = df.drop(columns=['Cognitive Load'])
y = df['Cognitive Load']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

X_train_reshaped = X_train[..., np.newaxis]
X_test_reshaped = X_test[..., np.newaxis]


class TransformerBlock(layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=0.1):
        super(TransformerBlock, self).__init__()
        self.att = layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)
        self.ffn = tf.keras.Sequential([
            layers.Dense(ff_dim, activation='relu'),
            layers.Dense(embed_dim)
        ])
        self.layernorm1 = layers.LayerNormalization(epsilon=1e-6)
        self.layernorm2 = layers.LayerNormalization(epsilon=1e-6)
        self.dropout1 = layers.Dropout(rate)
        self.dropout2 = layers.Dropout(rate)

    def call(self, inputs, training=None, **kwargs):
        attn_output = self.att(inputs, inputs, training=training)
        attn_output = self.dropout1(attn_output, training=training)
        out1 = self.layernorm1(inputs + attn_output)
        ffn_output = self.ffn(out1)
        ffn_output = self.dropout2(ffn_output, training=training)
        return self.layernorm2(out1 + ffn_output)

def masked_autoencoder(input_shape, num_heads, ff_dim, num_transformer_blocks, embed_dim):
    inputs = layers.Input(shape=input_shape)
    x = layers.Conv1D(embed_dim, 1, activation='relu')(inputs)

    for _ in range(num_transformer_blocks):
        x = TransformerBlock(embed_dim, num_heads, ff_dim)(x)

    # Prediction Head for Masked Autoencoder
    x = layers.Flatten()(x)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.Dropout(0.2)(x)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.2)(x)
    outputs = layers.Dense(X_train.shape[1])(x)  # Size equal to dimension of input features
    model = tf.keras.Model(inputs=inputs, outputs=outputs)
    return model
def downstream_classification_model(input_shape, num_heads, ff_dim, num_transformer_blocks, embed_dim):
    inputs = layers.Input(shape=input_shape)
    x = layers.Conv1D(embed_dim, 1, activation='relu')(inputs)

    for _ in range(num_transformer_blocks):
        x = TransformerBlock(embed_dim, num_heads, ff_dim)(x)
    x = layers.GlobalAveragePooling1D()(x)  # Pooling layer added
    x = layers.Dense(128, activation='relu')(x)  # Increased layer size
    x = layers.Dropout(0.3)(x)  # Increased dropout rate
    x = layers.Dense(64, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(1, activation='sigmoid')(x)  # Output for binary classification
    model = tf.keras.Model(inputs=inputs, outputs=outputs)
    return model
num_heads = 8
ff_dim = 128 
num_transformer_blocks = 8 
maxlen = X_train.shape[1]
embed_dim = 128  

# Build pre-training model
pre_training_model = masked_autoencoder((maxlen, 1), num_heads, ff_dim, num_transformer_blocks, embed_dim)
pre_training_model.compile(optimizer='adam', loss='mae')
history_pre_training = pre_training_model.fit(X_train_reshaped, X_train, epochs=30, batch_size=64)
downstream_model = downstream_classification_model((maxlen, 1), num_heads, ff_dim, num_transformer_blocks, embed_dim)
lr_scheduler = tf.keras.optimizers.schedules.CosineDecay(initial_learning_rate=1e-4, decay_steps=10000)
early_stopping = callbacks.EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True)

downstream_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr_scheduler), loss='binary_crossentropy', metrics=['accuracy'])
history_downstream = downstream_model.fit(X_train_reshaped, y_train, epochs=30, batch_size=64, validation_split=0.1, callbacks=[early_stopping])

# Evaluate the model
test_loss, test_accuracy = downstream_model.evaluate(X_test_reshaped, y_test)
print(f"Test accuracy: {test_accuracy}")
```

```python
import matplotlib.pyplot as plt

# Plotting training loss for pre-training
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history_pre_training.history['loss'], label='Pre-training Loss')
plt.title('Pre-training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Plotting training and validation loss for downstream task
plt.subplot(1, 2, 2)
plt.plot(history_downstream.history['loss'], label='Training Loss')
plt.plot(history_downstream.history['val_loss'], label='Validation Loss')
plt.title('Downstream Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay,classification_report
import tensorflow as tf
from tensorflow.keras import layers, callbacks
import matplotlib.pyplot as plt
y_pred_prob = downstream_model.predict(X_test_reshaped)
y_pred = (y_pred_prob > 0.5).astype(int)

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
disp.plot(cmap=plt.cm.Blues)
plt.show()
```

```python
report = classification_report(y_test, y_pred, target_names=['Class 0', 'Class 1'])
print(report)
```
